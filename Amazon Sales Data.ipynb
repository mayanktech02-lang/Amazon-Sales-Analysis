# Amazon Sales Data


## Description:

### This dataset contains information on 1K+ Amazon products, including their ratings, reviews, and other details.



# Features:

* product_id: Unique identifier for each product
* product_name: Name of the product
* category: Category of the product 
* discounted_price: Discounted price of the product
* actual_price: Actual price of the product
* discount_percentage: Percentage of discount for the product
* rating: Rating of the product (1-5)
* rating_count: Number of people who voted for the Amazon rating
* about_product: Description about the product
* user_id: ID of the user who wrote the review
* user_name: Name of the user who wrote the review
* review_id: ID of the user review
* review_title: Short review
* review_content: Long review
* img_link: Image link of the product
* product_link: Official website link of the product

```python
# Import some usefull libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/mayanktech02-lang/Amazon-Sales-Analysis/refs/heads/main/amazon.csv')
```

```python
df
```

```python
df.isnull().sum() # check null value in data
```

```python
df.dropna(inplace = True) # drop the null values, because only 2 palce found in rating_count column where null value present.
```

```python
df.duplicated().sum() #check duplicate value in data, in that dataset no duplicate data found
```

```python
df.info()
```

```python
df.discounted_price = df.discounted_price.str.replace('₹','') #remove the rupee sign from value
df.discounted_price = df.discounted_price.str.replace(',','')
```

```python
df.discounted_price= df.discounted_price.astype(float)
```

```python
df.actual_price = df.actual_price.str.replace('₹','')
df.actual_price = df.actual_price.str.replace(',','')
```

```python
df.actual_price = df.actual_price.astype(float)
```

```python
df.discount_percentage = df.discount_percentage.str.replace('%','')
df.discount_percentage = df.discount_percentage.str.replace(',','')
```

```python
df.discount_percentage = df.discount_percentage.astype(int)
```

```python
df = df[df["rating"] != ''] # check the data where the string value 
```

```python
df.rating = df.rating.str.replace('|','')
```

```python
df.rating = df.rating.astype(float)
```

```python
df.rating_count = df.rating_count.str.replace(',','')
```

```python
df.rating_count = df.rating_count.astype(float)
```

```python
df.info()
```

```python
cat_cols = df['category'].str.split('|', expand=True)

print(cat_cols.head())
```

```python
cat_cols.columns = [f'category_{i+1}' for i in range(cat_cols.shape[1])]
```

```python
df.drop(['category'], axis = 1, inplace= True)
```

```python
df = pd.concat([df, cat_cols], axis=1)
```

```python
pd.reset_option('all')
```

# 1 What is the average rating for each product categoryk

```python
df.groupby('category')['rating'].mean().sort_values(ascending=False)
```

```python
pivot_table = df.pivot_table(
    values='rating',
    columns='category',
    aggfunc='mean'
)
```

```python
plt.figure()
sns.heatmap(
    pivot_table,
    annot=True,
    fmt='.1f',
    cmap='coolwarm'
)

plt.title('Average product rating by category')
plt.show()
```

# 2. What are the top rating_count products by category

```python
top_products = df.loc[
    df.groupby('category')['rating_count'].idxmax(),
    ['category', 'product_name', 'rating_count']
]
```

```python
top_products.sort_values(by = 'rating_count', ascending=False)
```

# 4. How does the average discount percentage vary across categories

```python
plt.figure(figsize=(15,8))
df.groupby('category')['discount_percentage'].mean().plot()
plt.title('average discount percentage vary across categories')
plt.show()
```

# 5. What are the most popular product names

```python
popular_products = df[['product_name', 'rating_count']].sort_values(
    by='rating_count',
    ascending=False
)

popular_products.head(10)
```

```python

top10 = popular_products.head(10)

plt.figure(figsize=(10,5))
plt.barh(top10['product_name'], top10['rating_count'])

plt.xlabel('Rating Count')
plt.ylabel('Product Name')
plt.title('Top 10 Popular Products')

plt.show()
```

# 5. What are the most popular product keywords

```python
import re
from collections import Counter

text = ' '.join(df['product_name'].astype(str))

# keep only letters and numbers
text = re.sub(r'[^a-zA-Z0-9 ]', '', text)

words = text.lower().split()

keyword_count = Counter(words)

print(keyword_count.most_common(20))
```

# 7. What is the correlation between discounted_price and rating

```python
correlation = df['discounted_price'].corr(df['rating'])

print(correlation)
```

```python
plt.figure(figsize=(8,5))

plt.scatter(df['discounted_price'], df['rating'])

plt.xlabel('Discounted Price')
plt.ylabel('Rating')
plt.title('Discounted Price vs Rating')

plt.show()
```

# 9. What are the Top 5 categories based on the highest ratings

```python
highest_rating = df.groupby('category')['rating'].mean().round(2).sort_values(ascending=False).head(5)
```

```python
highest_rating.plot(
    kind='bar',
    figsize=(8,5)
)

plt.xlabel('Category')
plt.ylabel('Average Rating')
plt.title('Top 5 Categories by Rating')

plt.xticks(rotation=45)

plt.show()
```
